```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#dane do wykres√≥w
x = np.linspace(-2*np.pi,2*np.pi,100)
y = np.sinc(x)
```

```python
fig,ax = plt.subplots(ncols=2,figsize=(12,8))
ax[0].plot(x,y)
ax[0].set_title('pojedyncza krzywa')
ax[1].plot(x,y)
ax[1].plot(x*2.0,y)
ax[1].set_title('dwie krzywe')
plt.show()
```

```python
fig2,ax2 = plt.subplots(ncols=2,figsize=(12,8))
ax2[0].plot(x,y)
ax2[0].set_xlim(left=-1,right=1)
ax2[0].plot(x+np.pi*0.5,y)
ax2[0].set_title('set_xlim(left=-1,right=1)')


ax2[1].plot(x,y)
ax2[1].set_xlim(left=-1,right=1)
ax2[1].plot(x+np.pi*0.5,y)
ax2[1].autoscale()
ax2[1].set_title('set_xlim(left=-1,right=1)\nautoscale()')
plt.show()
```
