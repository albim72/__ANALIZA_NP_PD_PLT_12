```python
import numpy as np
import matplotlib.pyplot as plt
```

# Przykład 1: wykres kołowy - liczba grantów dla uczelni z wybranych miast

```python
#dane do wykesu
labels = 'Warszawa','Kraków','Lublin','Katowice','Wrocław'
print(type(labels))
```

```python
sizes = [56,34,27,28,18]
explode = (0,0.1,0,0,0.2)
```

```python
fig,ax = plt.subplots()
fig.patch.set_facecolor("#D3D3D3")
ax.set_facecolor("beige")
ax.pie(sizes,explode=explode,labels=labels,autopct="%1.1f%%",shadow=True,startangle=90)
ax.axis("equal")
plt.show()
```

# Przzykład 2: Wykres typu słupkowego (bar) określający wartości przyposane dla kategorii (A-K)

```python
#dane do wykresu
categories = ["A","B","C","D","E","K"]
values = [23,45,67,12,41,78]

colors = ["red","blue","orange","green","#45AA23","#783BCC"]
```

```python
fig,ax = plt.subplots(figsize=(8,6))

#ustawienia
fig.patch.set_facecolor("#D3D3D3")
ax.set_facecolor("beige")

bars = plt.bar(categories,values,color=colors,edgecolor="black")

#dodawanie etykiet
for bar,value in zip(bars,values):
    plt.text(bar.get_x() + bar.get_width()/2,bar.get_height()-5,str(value),ha="center", color="white",fontsize=16)

plt.title("Przykład wykresu słupkowego",fontsize=20)
plt.xlabel("Kategorie",fontsize=14)
plt.ylabel("Wartości",fontsize=14)
plt.grid(axis='y',linestyle='--',alpha=0.6)
plt.show()
```
